```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Data banate hain (X = Parhai ke ghante, y = Marks)
# Note: X ko 2D array hona chahiye [[...]]
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]]) 
y = np.array([15, 28, 35, 48, 55, 65, 78, 90])

# 2. Model ko initialize aur train karte hain
model = LinearRegression()
model.fit(X, y) # Model yahan seekh raha hai (training)

# 3. Prediction karte hain
# Maslan: Agar koi 5.5 ghante parhe toh kitne marks ayenge?
hours_to_test = np.array([[5.5]])
predicted_score = model.predict(hours_to_test)

print(f"Agar aap 5.5 ghante parhenge, toh aapke marks taqreeban {predicted_score[0]:.2f} ayenge.")

# 4. Visualizing (Graph banana)
plt.scatter(X, y, color='blue', label='Asli Data') # Purana data points
plt.plot(X, model.predict(X), color='red', label='Regression Line') # Seekhi hui line
plt.title('Hours vs Marks (Simple Linear Regression)')
plt.xlabel('Study Hours')
plt.ylabel('Exam Marks')
plt.legend()
plt.show()
```
